```python
import re

basefname = 'shahnameh.warner.xml'

tmpfname = re.sub('\.xml','-tmp.xml',basefname)
outfname = re.sub('\.xml','-fixed.xml',basefname)

f = open(basefname)
tmpf = open(tmpfname,'w')

notedict = {}
refdict = {}

divlev = 0
curpage = 'notyet'
inintro = 0
curvolume = '0'
cursectn = 0
prevsectn = 0
noten = 0
curchunkn = 0
for l in f:
    m = re.search('<pb n="([^"]+)"',l)
    if(m):
        curpage = m[1]
        noten = 0

    l = re.sub('\s*$','',l)
    l = re.sub('\s+</note>','</note>',l)
    l = re.sub('<note xml:id="n.','<note n="',l)
    
    while(re.search('<note>',l)):
        noten = noten + 1
        l = re.sub('<note>','<note n="'+str(noten)+'">',l,1)
        #print(l)
    
    m = re.search('<milestone unit="volume" n="([0-9]+)"',l)
    if(m):
        curvolume = m[1]
        
    l = re.sub('target="n\.','n="',l)
    
    if(re.search('(<note|<ref)\s+(n=")([0-9]+")',l)):
        #print()
        #print(l)
        l = re.sub('(<note|<ref)\s+(n=")([0-9]+")','\g<1> \g<2>'+curpage + '.\g<3>',l)
        #print(l)
    if(re.search('(<note|<ref)\s+(n=")([0-9]+\.[0-9]+")',l)):
        l = re.sub('(<note|<ref)\s+(n=")([0-9]+\.[0-9]+")','\g<1> \g<2>'+curvolume + '.\g<3>',l)
    
    if(re.search('"(intro|argument)"',l)):
        inintro = 1
    if(re.search('</div>',l)):
        inintro = 0
        
    if(inintro == 0 ):
        l = re.sub('(<p>)([A-Z]|<q>)','\g<1>\n<lb/>\g<2>',l)
        if(re.search('^([A-Z]|<q>|\()',l) ):
            l = '<l/>' + l
        elif( not curpage == 'notyet' and not re.search('(\(|</lg>|<mile|<pb|</p>|<p>|<l|</div|<div|<note)',l) and l):
            print(curpage,l)
    else:
        l = re.sub('<lb/>','',l)
        #if(re.search('<lb/>',l)):
        #    print('lb!!',l)
        
        #print(l)
    m = re.search('<pb n="([^"]+)"',l)
    if(m):
        curpage = m[1]
    if(re.search('<div',l)):
        divlev = divlev + 1
        if(re.search('(zreign|"part")',l) or 1):
            if(re.search('"reign"',l) and 0):
                print('\n')
            #print(divlev,l)
    if(re.search('</div>',l)):
        divlev = divlev - 1
        
    m = re.search('subtype="reign" n="([^"]+)"',l)
    if(m):
        curreign = m[1]
        curchunkn = 0
       
       
    m = re.search('subtype="section" n="([0-9]+)"',l)
    if(m):
        cursect = m[1]
        cursectn = int(m[1])
        if(not cursectn == prevsectn + 1 and not cursectn == 1):
            print(curpage,'skip',prevsectn,cursectn)
        if(cursectn == 1):
            curchunkn = curchunkn + 1
        l = re.sub('subtype="section" n="([0-9]+)"','subtype="section" n="'+curreign+'.'+str(curchunkn)+'.\g<1>"',l)
        #print(curchunkn,cursect,l)
            
        prevsectn = cursectn
        
        
    if(re.search('^<note',l) and not re.search('</note>',l)):
        print(l)
        
    l = re.sub('(<xref n=")([0-9])','\g<1>Mohl \g<2>',l)
    print(l,file=tmpf)
    
    while(re.search('<ref n="([^"]+)"/>',l)):
        m = re.search('<ref n="([^"]+)"/>',l)
        refdict[m[1]] = 1
        l = re.sub('<ref n="([^"]+)"/>','',l,1)
    
    m = re.search('^<note n="([^"]+)">(.+)</note>',l)
    if(m):
        notedict[m[1]] = m[2]
        
f.close()  
tmpf.close()

for foo in refdict:
    if(not foo in notedict):
        print('nonote',foo)

for foo in notedict:
    if(not foo in refdict):
        print('noref',foo)
    
```

```python
warner2mohlx = {}
f = open('shahnameh.mohl-fixed.xml')
for l in f:
    cur2w = ''
    curid = ''
    m = re.search('<xref n="Warner\s+([^"]+)',l)
    if(m):
        cur2w = m[1]
    m = re.search('section" n="([^"]+)',l)
    if(m):
        curid = m[1]
    if(cur2w and curid):
        warner2mohlx[cur2w] = curid
    
f.close()

f = open(tmpfname)
outf = open(outfname,'w')
newpara = 0
qlevel = 0
lnum = 0
innote = 0
for l in f:
    l = re.sub('\s+$','',l)
    l = re.sub('<lb/>','<l/>',l)
    
    m = re.search('section" n="([^"]+)"',l)
    if(m and m[1] in warner2mohlx):
        if(not re.search('<head>',l)):
            print('nohead',l)
        subval = '<xref n="Mohl '+ warner2mohlx[m[1]] + '"/>'
        l = re.sub('<head>','<head>'+subval,l)
        #print(m[1],subval)
    
    if(re.search('(<p>|^</p>)',l)):
        newpara = 0
        
    #if(qlevel>0 ):
    #    newpara = 0

    #l = str(qlevel) + ':' + str(newpara) + l
    if(newpara and re.search('<l/>',l)):
        l = '</p><p>'  + '\n' + l
        #print('hit',l)
        newpara = 0
    
    if(re.search('(<l/>)',l)):
        ql = l
        ql,qplus = re.subn('<q>','',ql)
        qlevel = qlevel + qplus
        ql,qminus = re.subn('</q>','',ql)
        qlevel = qlevel - qminus
        #l = re.sub('(<l/>)','\g<1>' + str(qlevel),l)
    
    
    if(not qlevel and re.search('([!\?\.])(</q></q>|</q>)$',l)):
        #print(qlevel,l)
        newpara = 1
    
    while(re.search('<ref n="([^"]+)"',l)):
        
        m = re.search('<ref n="([^"]+)"',l)
        curnote = notedict[m[1]]
        #print(curnote)
        l = re.sub('<ref n="([^"]+)"','<note n="\g<1>">'+curnote+'</note>',l,1)
    
    l = re.sub('<lb/>','<l/>',l)
    l = re.sub('^<note[^>]*>.+</note>','',l)
    
    if(re.search('type="section"',l)):
        lnum = 0
        
    if(re.search('<l/>',l)):
        lnum = lnum + 1
        l = re.sub('<l/>','<l n="'+str(lnum)+'"/>',l)
    if(re.search('subtype="intro"',l)):
        innote = 1
    if(re.search('</div>',l)):
        innote = 0
    if(not innote and re.search('<l>',l)):
        lnum = lnum + 1
        l = re.sub('<l>','<l n="'+str(lnum)+'">',l)

        
    print(l,file=outf)

f.close()
outf.close()
```

```python
import re
mohlaligns = {}

f = open('warner-mohl-secs.tsv')
for l in f:
    l = re.sub('\s+$','',l)
    
    args = l.split('\t')
    if(not len(args)==5):
        print(len(args),l)
    curreign = args[4]
    curmohl = args[2]
    if(re.search('<xref',curmohl)):
        #print(l)
        continue
    curkey = curmohl + '\t' + curreign
    if(curkey in mohlaligns):
        print('dupe',curkey)
    if(args[0]):
        mohlaligns[curkey] = args[0]
    #print(len(args))

f.close()
```

```python
#add section ids

import re
matchedmohl = {}
curreign = ''
curpart = 0
cursection = 0


f = open('shahnameh.mohl.xml')
outf = open('shahnameh.mohl-fixed.xml','w')
for l in f:
    l = re.sub('\s+$','',l)
    m = re.search('reign" n="([^"]+)"',l)
    if(m):
        curreign = m[1]
        cursection = 0
        curpart = 1
        
    m = re.search('subtype="part"\s+n="([0-9]+)">',l)
    if(m):
        curpart = int(m[1])
    m = re.search('subtype="section">',l)
    if(m):
        
        cursection = cursection + 1
        l = re.sub('subtype="section">','subtype="section" n="'+curreign+'.'+str(curpart)+'.'+str(cursection)+'">',l)
    m = re.search('<head>(.+)</head>',l)
    if(m):
        curhead = m[1]
        headkey = curhead + '\t' + curreign
        if(headkey in mohlaligns):
            subval = '<xref n="'+mohlaligns[headkey]+'"/>' + curhead
            l = re.sub('<head>.+</head>','<head>'+subval+'</head>',l)
            matchedmohl[headkey] = curhead
    l = re.sub('(<xref n=")(pre|[0-9])','\g<1>Warner \g<2>',l)
    print(l,file=outf)

f.close()
outf.close()

for foo in mohlaligns:
    if( not foo in matchedmohl):
        print('missed',foo)
```

```python
import re

warn2mohl = {}
mohl2warn = {}
warnids = {}
f = open('shahnameh.warner-fixed.xml')
for l in f:
    l = re.sub('(<xref n=")([0-9])','\g<1>Mohl \g<2>',l)
    
    m = re.search('section" n="([^"]+)"',l)
    if(m):
        warnids[m[1]] = 1
    m = re.search('<xref n="Mohl\s+([^"]+)"',l)
    if(m):
        warn2mohl[m[1]] = 1
f.close()

f = open('shahnameh.mohl-fixed.xml')
outf = open('mohl2warner','w')

i = 0
for l in f:
    m = re.search('<xref n="Warner\s+([^"]+)"',l)
    if(m):
        curm2 = m[1]
        if(curm2 in mohl2warn):
            print('dupe m2w',curm2)
        mohl2warn[curm2] = 1
    m = re.search('subtype="section" n="([^"]+)"',l)
    if(not m):
        continue
    curmohl = m[1]
    #print(curmohl)
    
    m = re.search('<xref n="Warner\s+([^"]+)"',l)
    if(not m):
        if(not curmohl in warn2mohl):
            i = i + 1
            print(i,'noxref',l)
    


outf.close()
f.close()

for foo in warnids:
    if(not foo in mohl2warn):
        print('mohl2warn missing',foo)
```
