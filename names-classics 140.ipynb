```python
import re

f = open('shahnameh.warn02.xml')

people = {}
places = {}
things = {}

for l in f:
    #print(l,end='')
    while(re.search('<persName[^>]*>([^<]+)',l)):
        m = re.search('<persName[^>]*>([^<]+)',l)
        curn = m[1]
        if(curn in people):
            people[curn] = people[curn] + 1
        else:
            people[curn] =  1
        l = re.sub('<persName[^>]*>([^<]+)',' ',l,1)

    while(re.search('<rs type="([^"]+)"[^>]*>([^<]+)',l)):
        m = re.search('<rs type="([^"]+)"[^>]*>([^<]+)',l)
        curtype = m[1]
        curn = m[2]
        curkey = m[2] + ' ' + m[1]
        if(curkey in things):
            things[curkey] = things[curkey] + 1
        else:
            things[curkey] =  1
        l = re.sub('<rs type="([^"]+)"[^>]*>([^<]+)',' ',l,1)

    while(re.search('<placeName[^>]*>([^<]+)',l)):
        m = re.search('<placeName[^>]*>([^<]+)',l)
        curn = m[1]
        if(curn in places):
            places[curn] = places[curn] + 1
        else:
            places[curn] =  1
        l = re.sub('<placeName[^>]*>([^<]+)',' ',l,1)
f.close()

i = 0
maxv = 30
print("Key figures in the Shahnameh (Warner transliteration)")
print("Basic information for (most) of these is available in the Warner Index to volume 2\nhttps://archive.org/details/shahnama02firduoft/page/412/mode/2up")
print()
print("People")
for foo in sorted(people,key=people.get,reverse=True):
    print(people[foo],foo)
    i = i + 1
    if( i > maxv):
        break

print()
print('Places')
i = 0
for foo in sorted(places,key=places.get,reverse=True):
    print(places[foo],foo)
    i = i + 1
    if( i > maxv):
        break

print()
print('Other')
i = 0
for foo in sorted(things,key=things.get,reverse=True):
    print(things[foo],foo)
    i = i + 1
    if( i > maxv):
        break
```
